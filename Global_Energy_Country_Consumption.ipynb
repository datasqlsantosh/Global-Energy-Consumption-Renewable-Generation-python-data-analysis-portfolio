```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Country_Consumption = pd.read_csv('/content/sample_data/Country_Consumption_TWH.csv')
print(Country_Consumption)
```

```python
# Bar Plot
plt.figure(figsize=(12, 6))
sns.barplot(data=Country_Consumption, palette="viridis")
plt.title('Comparison of Country-wise Data', fontsize=16)
plt.xlabel('Country', fontsize=14)
plt.ylabel('Values', fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Plotting using Seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x='China', y='Year', data=Country_Consumption, orient='horizontal' ,palette='viridis')

# Adding labels and title
plt.xlabel('China')
plt.ylabel('Year')
plt.title('Energy Consumption China')

plt.tight_layout()
plt.show()
```

```python
# Box Plot
plt.figure(figsize=(12, 6))
sns.boxplot(data=Country_Consumption, palette="Set3")
plt.title('Distribution of Country-wise Data', fontsize=16)
plt.xlabel('Country', fontsize=14)
plt.ylabel('Values', fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(Country_Consumption, cmap="YlGnBu", annot=True, fmt="d")
plt.title('Heatmap of Country-wise Data', fontsize=16)
plt.xlabel('Country', fontsize=14)
plt.ylabel('Year', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# Scatter Plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=Country_Consumption, palette="tab10", s=100)
plt.title('Scatter Plot of Country-wise Data', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Values', fontsize=14)
plt.legend(title='Country', title_fontsize='13', loc='upper left', bbox_to_anchor=(1, 1))
plt.tight_layout()
plt.show()
```
