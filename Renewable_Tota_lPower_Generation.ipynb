```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
RenewablesTotalPowerGeneration = pd.read_csv('/content/sample_data/renewablesTotalPowerGeneration.csv')
print(RenewablesTotalPowerGeneration)
```

```python
print(RenewablesTotalPowerGeneration.head())
```

```python
print(RenewablesTotalPowerGeneration.describe())
```

```python
# Check for missing values
print(RenewablesTotalPowerGeneration.isnull().sum())
```

```python
#Pandas tail() method is used to return last n (5 by default) rows of a data frame or series.
print(RenewablesTotalPowerGeneration.tail())
```

```python
#return size dimensions,and shape respectively
print('RenewablesTotalPowerGeneration',RenewablesTotalPowerGeneration.shape)
```

```python
sns.barplot(x='Mode of Generation', y='Contribution (TWh)', data=RenewablesTotalPowerGeneration, palette='viridis')

plt.title('Mode of Generation Contribution')
plt.xlabel('Mode of Generation')
plt.ylabel('Contribution (TWh)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Plotting a Pie Chart
plt.figure(figsize=(9, 9))
plt.pie(RenewablesTotalPowerGeneration['Contribution (TWh)'], labels=RenewablesTotalPowerGeneration['Mode of Generation'], autopct='%1.1f%%', startangle=140, colors=sns.color_palette('viridis', len(RenewablesTotalPowerGeneration)))
plt.title('Mode of Generation Contribution')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# Horizontal Bar Plot
plt.figure(figsize=(10, 6))
sns.barplot(x='Contribution (TWh)', y='Mode of Generation', data=RenewablesTotalPowerGeneration, palette='viridis')
plt.title('Mode of Generation Contribution')
plt.xlabel('Contribution (TWh)')
plt.ylabel('Mode of Generation')
plt.tight_layout()
plt.show()
```
