```python
# Stacked Area Chart
plt.figure(figsize=(12, 6))
Continent_Consumption.plot(kind='area', stacked=True)
plt.title('Cumulative Energy Production Trends by Region')
plt.xlabel('Year')
plt.ylabel('Production (TWh)')
plt.legend(title='Region')
plt.grid(True)
plt.show()
```

```python
# Plotting the sorted data
plt.figure(figsize=(10, 6))
sns.barplot(x='World', y='Year', data=Continent_Consumption, orient='horizontal', palette='coolwarm')
plt.title('Highest Year of World Energy Consumption')
plt.xlabel('World')
plt.ylabel('Year')
plt.show()
```

```python
# Plotting the sorted data
plt.figure(figsize=(10, 6))
sns.barplot(x='North America', y='Year', data=Continent_Consumption, orient='horizontal', palette='coolwarm')
plt.title('Highest Year of North_America Energy Consumption')
plt.xlabel('North America')
plt.ylabel('Year')
plt.show()
```

```python
# Plotting using Seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x='Latin America', y='Year', data=Continent_Consumption, orient='horizontal' ,palette='viridis')

# Adding labels and title
plt.xlabel('Latin America')
plt.ylabel('Year')
plt.title('Energy Consumption Latin America Region in 1990')

plt.tight_layout()
plt.show()
```

```python
# Bar plot for Energy by Region
plt.bar(Continent_Consumption.columns[2:], Continent_Consumption.values[0][2:], color='skyblue')
plt.title('Energy by Region in 1990')
plt.xlabel('Region')
plt.ylabel('Energy')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Area Plot
Continent_Consumption.plot.area(x='Year', y=['OECD', 'BRICS', 'Asia', 'CIS'])
plt.title('Energy Consumption')
plt.xlabel('Year')
plt.ylabel('Energy Consumption')
plt.show()
```
