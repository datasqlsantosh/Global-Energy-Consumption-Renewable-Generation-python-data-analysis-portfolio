```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
CountriesPowerGeneration = pd.read_csv('/content/sample_data/top20CountriesPowerGeneration.csv')
print(CountriesPowerGeneration)
```

```python
print(CountriesPowerGeneration.head())
```

```python
print(CountriesPowerGeneration.describe())
```

```python
# Check for missing values
print(CountriesPowerGeneration.isnull().sum())
```

```python
#Pandas tail() method is used to return last n (5 by default) rows of a data frame or series.
print(CountriesPowerGeneration.tail())
```

```python
#return size dimensions,and shape respectively
print('Countries Power Generation',CountriesPowerGeneration.shape)
```

```python
# Drop the 'Total' column
CountriesPowerGeneration.drop(columns=['Total (TWh)'], inplace=True)
print(CountriesPowerGeneration.head())
```

```python
CountriesPowerGeneration.plot(kind='bar', stacked=False, figsize=(10, 6), colormap='viridis')

plt.title('Energy Production by Country and Source')
plt.xlabel('Country')
plt.ylabel('Production (TWh)')
plt.xticks(rotation=0)
plt.legend(title='Energy Source')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Create a bar plot using seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x='Country', y='Hydro(TWh)', data=CountriesPowerGeneration , palette='viridis')

# Adding labels and title
plt.xlabel('Country')
plt.ylabel('Hydro(TWh)')
plt.title('Country of most hydrogen')

# Rotating x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Display the plot
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Hydro(TWh)', y='Country', data=CountriesPowerGeneration, orient='horizontal', palette='coolwarm')
plt.title('Country of most Hydrogen')
plt.xlabel('Country')
plt.ylabel('Hydro(TWh)')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Biofuel(TWh)', y='Country', data=CountriesPowerGeneration, orient='horizontal', palette='coolwarm')
plt.title('Country of most Biofuel')
plt.xlabel('Country')
plt.ylabel('Biofuel(TWh)')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Solar PV (TWh)', y='Country', data=CountriesPowerGeneration, orient='horizontal', palette='coolwarm')
plt.title('Country of most Solar PV')
plt.xlabel('Country')
plt.ylabel('Solar PV (TWh)')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Geothermal (TWh)', y='Country', data=CountriesPowerGeneration, orient='horizontal', palette='coolwarm')
plt.title('Country of most Geothermal')
plt.xlabel('Country')
plt.ylabel('Geothermal (TWh)')
plt.show()
```

```python
# Stacked Area Chart
plt.figure(figsize=(12, 6))
CountriesPowerGeneration.plot(kind='area', stacked=True)
plt.title('Energy Production Trends by Country')
plt.xlabel('Year')
plt.ylabel('Production (TWh)')
plt.legend(title='Country')
plt.grid(True)
plt.show()
```
