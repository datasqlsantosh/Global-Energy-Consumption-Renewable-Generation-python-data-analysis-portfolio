```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
NonrenewablesTotalPowerGeneration = pd.read_csv('/content/sample_data/nonRenewablesTotalPowerGeneration.csv')
print(NonrenewablesTotalPowerGeneration)
```

```python
print(NonrenewablesTotalPowerGeneration.head())
```

```python
print(NonrenewablesTotalPowerGeneration.describe())
```

```python
# Check for missing values
print(NonrenewablesTotalPowerGeneration.isnull().sum())
```

```python
#Pandas tail() method is used to return last n (5 by default) rows of a data frame or series.
print(NonrenewablesTotalPowerGeneration.tail())
```

```python
#return size dimensions,and shape respectively
print('NonrenewablesTotalPowerGeneration',NonrenewablesTotalPowerGeneration.shape)
```

```python
sns.barplot(x='Mode of Generation', y='Contribution (TWh)', data=NonrenewablesTotalPowerGeneration, palette='viridis')

plt.title('Mode of Generation Contribution')
plt.xlabel('Mode of Generation')
plt.ylabel('Contribution (TWh)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Plotting a Pie Chart
plt.figure(figsize=(10, 10))
plt.pie(NonrenewablesTotalPowerGeneration['Contribution (TWh)'], labels=NonrenewablesTotalPowerGeneration['Mode of Generation'], autopct='%1.1f%%', startangle=140, colors=sns.color_palette('viridis', len(NonrenewablesTotalPowerGeneration)))
plt.title('Mode of Generation Contribution')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# Horizontal Bar Plot
plt.figure(figsize=(10, 6))
sns.barplot(x='Contribution (TWh)', y='Mode of Generation', data=NonrenewablesTotalPowerGeneration, palette='viridis')
plt.title('Mode of Generation Contribution')
plt.xlabel('Contribution (TWh)')
plt.ylabel('Mode of Generation')
plt.tight_layout()
plt.show()
```
